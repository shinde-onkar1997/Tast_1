## Hypothesis Testing

```python
df1.groupby(df1['Temperature (C)']).mean()
```

```python
df1.groupby(df1['Humidity']).mean()
```

```python
temperature = df1[(df1['Temperature (C)'] == 'Temperature')]
humidity = df1[(df1['Humidity'] == 'Humidity')]
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(temperature, humidity, equal_var = True)
```
